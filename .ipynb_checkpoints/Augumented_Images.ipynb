```python
import os
import shutil
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img

train_dir = "/Users/ligiahelena/Documents/Kaggle/Dataset/train"
augmented_train_dir = "/Users/ligiahelena/Documents/Kaggle/Dataset/augmented_train"

os.makedirs(augmented_train_dir, exist_ok=True)

#Class folders and count images
class_counts = {cls: len(os.listdir(os.path.join(train_dir, cls))) for cls in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, cls))}

#Threshold for minority classes
minority_threshold = 100
minority_classes = [cls for cls, count in class_counts.items() if count < minority_threshold]

# Data augmentation settings
datagen = ImageDataGenerator(
    rotation_range=5,       
    width_shift_range=0.02,  
    height_shift_range=0.02, 
    shear_range=0.05,        
    zoom_range=0.05,         
    brightness_range=[0.9, 1.1],
    horizontal_flip=True,    
    vertical_flip=False      
)

#Process each class
for class_name in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_name)
    
    #Skip non-directory files (I was getting an error for .DS_Store)
    if not os.path.isdir(class_path):
        continue
    
    augmented_class_path = os.path.join(augmented_train_dir, class_name)
    os.makedirs(augmented_class_path, exist_ok=True)  

    #Copy original images
    for img_name in os.listdir(class_path):
        shutil.copy(os.path.join(class_path, img_name), os.path.join(augmented_class_path, img_name))
    
    #skip augmentation if not minority
    if class_name not in minority_classes:
        continue

    augmentation_factor = 1.5  #Increase minority classes by 50%
    minimum_threshold = 200  #Minimum number of images per class

    #Augment minority class images
    num_original = class_counts[class_name]
    target_count = max(int(num_original * augmentation_factor), minimum_threshold)
    num_to_generate = target_count - num_original

    print(f"Augmenting {class_name}: {num_original} images → Generating {num_to_generate} more")

    image_files = os.listdir(class_path)
    
    for i in range(num_to_generate):
        img_name = np.random.choice(image_files)  #Randomly pick an image
        img_path = os.path.join(class_path, img_name)
        
        #Augment the image
        img = load_img(img_path)
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        
        aug_iter = datagen.flow(img_array, batch_size=1)
        aug_img = next(aug_iter)[0].astype('uint8')
        
        #Save it
        aug_img_pil = array_to_img(aug_img)
        aug_img_pil.save(os.path.join(augmented_class_path, f"aug_{i}_{img_name}"))

print(f"Augmentation complete! Augmented images stored in: {augmented_train_dir}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to the augmented dataset
augmented_train_dir = "/Users/ligiahelena/Documents/Kaggle/Dataset/augmented_train"

# Count images per class in the augmented dataset
aug_label_counts = {
    class_name: len(os.listdir(os.path.join(augmented_train_dir, class_name)))
    for class_name in os.listdir(augmented_train_dir)
    if os.path.isdir(os.path.join(augmented_train_dir, class_name))
}

# Convert to DataFrame
aug_label_df = pd.DataFrame(aug_label_counts.items(), columns=["label", "count"]).sort_values(by="count", ascending=False)

# Plot the distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=aug_label_df["label"], y=aug_label_df["count"])
plt.xticks(rotation=90)
plt.title("Distribution of Skin Condition Labels (Augmented Train)")
plt.xlabel("Skin Condition")
plt.ylabel("Image Count")
plt.show()
```
