```python
import shutil
import hashlib
import os

# Define the base path (adjust this to your actual directory structure)
base_path = 'C:/Users/Manik/Desktop/BTT Project/BTAI-AJL-Team-2/bttai-ajl-2025'

# Define paths for train, extended, and combined datasets
train_path = os.path.join(base_path, 'train', 'train')
extended_path = os.path.join(base_path, 'extended_dataset')
combined_path = os.path.join(base_path, 'combined_dataset')

# Create the combined_path directory if it doesn't exist
os.makedirs(combined_path, exist_ok=True)

# Get all unique categories from both train and extended datasets
categories_train = {d for d in os.listdir(train_path) if os.path.isdir(os.path.join(train_path, d))}
categories_extended = {d for d in os.listdir(extended_path) if os.path.isdir(os.path.join(extended_path, d))}
all_categories = categories_train.union(categories_extended)

# Function to compute MD5 hash of a file
def compute_md5(file_path):
    """Compute the MD5 hash of a file to identify its content."""
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

# Function to copy files with duplicate checking and conflict handling
def copy_files_with_duplicate_check(source_folder, dest_folder, existing_hashes):
    """
    Copy files from source to destination, skipping duplicates based on content (MD5 hash).
    
    Args:
        source_folder (str): Path to the source folder containing files.
        dest_folder (str): Path to the destination folder.
        existing_hashes (set): Set of MD5 hashes of files already copied.
    """
    for filename in os.listdir(source_folder):
        source_file = os.path.join(source_folder, filename)
        if os.path.isfile(source_file):  # Ensure it's a file, not a directory
            # Compute the MD5 hash of the file
            file_hash = compute_md5(source_file)
            
            # Skip if the file content is already in the destination
            if file_hash in existing_hashes:
                print(f"Skipping duplicate file: {source_file} (hash: {file_hash})")
                continue
            
            # Handle filename conflicts (same name, different content)
            dest_file = os.path.join(dest_folder, filename)
            if os.path.exists(dest_file):
                base, ext = os.path.splitext(filename)
                i = 1
                while os.path.exists(os.path.join(dest_folder, f"{base}_{i}{ext}")):
                    i += 1
                dest_file = os.path.join(dest_folder, f"{base}_{i}{ext}")
            
            # Copy the file and update the hash set
            try:
                shutil.copy2(source_file, dest_file)
                print(f"Copied {source_file} to {dest_file}")
                existing_hashes.add(file_hash)
            except Exception as e:
                print(f"Error copying {source_file} to {dest_file}: {e}")

# Combine images for each category
for category in all_categories:
    # Create the category subfolder in combined_dataset
    category_path = os.path.join(combined_path, category)
    os.makedirs(category_path, exist_ok=True)
    
    # Track hashes of files already copied to this category
    existing_hashes = set()
    
    # Copy from train_path if the category exists
    train_category_path = os.path.join(train_path, category)
    if os.path.exists(train_category_path):
        copy_files_with_duplicate_check(train_category_path, category_path, existing_hashes)
    
    # Copy from extended_path if the category exists
    extended_category_path = os.path.join(extended_path, category)
    if os.path.exists(extended_category_path):
        copy_files_with_duplicate_check(extended_category_path, category_path, existing_hashes)

print("Dataset combination complete with duplicate checking!")
```

```python
import pandas as pd

# Define file paths
train_path = 'C:/Users/Manik/Desktop/BTT Project/BTAI-AJL-Team-2/bttai-ajl-2025/train.csv'
filtered_train_path = 'C:/Users/Manik/Desktop/BTT Project/BTAI-AJL-Team-2/bttai-ajl-2025/filtered_train.csv'
combined_path = 'C:/Users/Manik/Desktop/BTT Project/BTAI-AJL-Team-2/bttai-ajl-2025/both_combined.csv'

# Step 1: Load train.csv with all columns and rows
train_df = pd.read_csv(train_path)

# Step 2: Define the columns to select from filtered_train.csv
selected_columns = [
    'md5hash',
    'fitzpatrick_scale',
    'fitzpatrick_centaur',
    'label',
    'nine_partition_label',
    'three_partition_label',
    'qc'
]

# Load filtered_train.csv and select only the specified columns
filtered_train_df = pd.read_csv(filtered_train_path)[selected_columns]

# Step 3: Concatenate the DataFrames vertically
combined_df = pd.concat([train_df, filtered_train_df], ignore_index=True)

# Step 4: Drop duplicates based on 'md5hash', keeping the first occurrence
combined_df = combined_df.drop_duplicates(subset='md5hash', keep='first')

# Step 5: Save the combined DataFrame to a new CSV file
combined_df.to_csv(combined_path, index=False)

print(f"Combined CSV saved to {combined_path}")
```
